```python
#import random and sys 
import random,sys

#Create a word list with 10 different word
word_list = ["hello", "world", "this", "place", "is", "crutial", "to", "live","longword", "morelongword"]
# Create an alphabet
alp = "abcdefghijklmnopqrstuvwxyz"
alphabet = list(alp)
#check your secret word lenght
guess_word = []
secret_word = random.choice(word_list)
word_lenght = len(secret_word)
letter_storage = []

#define a function that prints your current word
def print_word_to_guess(letters):
    """Utility function to print the current word to guess"""
    print("Word to guess: {0}".format(" ".join(letters)))
    
#define a function that prints players current guess and total guess chances 
def print_guess_taken(current,total):
    """Prints how many chances the player has used"""
    print(f"You are on guess {current}/{total}.")

#define a function that says the game starts and ask players name
def game_start():
    print("Hello Mate!! \n")
    
    while True:
        name = input("What is your name?").strip()
        if name != "":
            print("Hello " + name)
            break
        else:
            print("Your name can't be blank.\n Please enter your name")
# define a function to Ask users preference to play
def ask_user_to_play():
    """Ask user if they want to play"""
    print("Well, that's perfect moment to play some Hangman!\n")
    while True:
        x = input("Would you like to play? Yes or no?\n").lower()
        if x == "yes" or x == "y":
            print("Game Started.")
            break
        if x == "no" or x == "n":
            sys.exit("That's a shame. May be next time.")
        else:
            print("Please only enter yes or no")
            
#define a function that prepares your secret word
def prepare_secret_word():
    for char in secret_word:
        guess_word.append("-")
    print("Okey your word lenght is", word_lenght ," characters")
    print("Be aware that You can enter only 1 letter from a-z\n\n")
    print_word_to_guess(guess_word)
#define a function that inform you about your guessing.
def guessing():
    guess_taken=1
    max_guess=6
    while guess_taken <= max_guess:
        guess = input("Pick a letter\n").lower()
        if not guess in alphabet:
            print("Enter a letter from a-z alphabet.")
        elif guess in letter_storage:
            print("You already guessed that letter, Please choose new one.")
        else:
            letter_storage.append(guess)
            if guess in secret_word:
                print("Your guessing is correct.")
                for i in range(0,word_lenght):
                    if secret_word[i]== guess:
                        guess_word[i] = guess
                print_word_to_guess(guess_word)
                print_guess_taken(guess_taken,max_guess)
                if not "-" in guess_word:
                    print("You won!")
                    print("Game over!")
                    break
            else:
                if guess_taken==0:                                                
                  print("""                                               



                              
                                   """)
                elif guess_taken==1:
                  print("""



                              
                              ____""")
                elif guess_taken==2:
                  print("""
                                 
                                |
                                |
                                |
                                |
                              __|__""")
                elif guess_taken==3:
                  print("""
                                 ___
                                |   
                                |   
                                |
                                |
                              __|__""")
                elif guess_taken==4:
                  print("""
                                 ___
                                |   |
                                |   
                                |  
                                |
                              __|__""")
                elif guess_taken==5:
                  print("""
                                 ___
                                |   |
                                |   O
                                |       
                                |    
                              __|__""")
      
                elif guess_taken==6:
                  print("""
                                 ___
                                |   |
                                |   O
                                |  /|\     
                                |  / \     
                              __|__""")
                  print(f" Sorry Mate, You lost :<! The secret word was: {secret_word}")
                  break
                
                print("The letter is not in the word. Try Again!")
                guess_taken += 1
                print_guess_taken(guess_taken, max_guess)

if __name__ == "__main__":
    game_start()
    ask_user_to_play()
    prepare_secret_word()
    guessing()
#Call your function to Run your script 
```
